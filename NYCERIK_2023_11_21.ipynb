```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/Salaries.csv')
```

```python
df.describe
```

```python
df.head()
```

Rename discipline

```python
df=df.rename(columns={'discipline': 'degree'})
```

```python
df.head()
```

created salary subset with stats

```python
df_salary=df['salary']
```

```python
df_salary.describe()
```

Grouped genders for median salary

```python
df.groupby(['sex'])['salary'].median()
```

Grouped rank for median salary

```python
df.groupby(['rank'])['salary'].mean()
```

Gender visualization by salary

```python
df.groupby(['sex'])['salary'].median().plot(kind='bar')
```

Rank visualization by mean salary

```python
df.groupby(['rank'])['salary'].mean().plot(kind="pie")
```
