```python
#import pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import csv dataset
df=pd.read_csv(' https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv', index_col=False)
```

```python
#confirm import by checking headers and shape
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
# Export the DataFrame to an Excel file
df.to_excel('output.xlsx', index=False)
```

```python
#Check for missing data - total dataframe
df.isnull().sum()
```

```python
#Replace na value with zero(0)
df.fillna(value='0')
```

```python
#Rename School Level to School Class Years
df=df.rename(columns={'School Level': 'School Class Years'})
```

```python
#verify column name change
df.head()
```

```python
#School No. 1 Astor Collegiate Academy - Remove columns not necessary for analysis
df.drop(['School DBN', 'School Class Years','Total Tested','Number Scoring Below 65', 'Percent Scoring Below 65', 'Number Scoring 65 or Above', 'Percent Scoring 65 or Above', 'Number Scoring 80 or Above', 'Percent Scoring 80 or Above', 'Number Scoring CR', 'Percent Scoring CR'], axis=1, inplace=True)
```

```python
#Confirm columns removed
df.head()
```

```python
#Choose two schools to compare based upon simliar type of school (Collegiate High School) similar total tested size ,one in Queens and one in the Bronx
dfschool1=df.loc[df['School Name'].isin(['Astor Collegiate Academy', 'Rockaway Collegiate High School'])]
```

```python
#Confirm school were selected form population - top
dfschool1.head()
```

```python
#Confirm school were selected form population - bottom
dfschool1.tail()
```

```python
#Create subset of Regents Exam = Common Core English
dfexam=dfschool1[dfschool1['Regents Exam'] == 'Common Core English']
```

```python
#Confirm rows were filtered for exam type
dfexam.head()
```

```python
#confirm new subset size
dfexam.shape
```

```python
#filter out by year
dfexamyear=dfexam[dfexam['Year'] == 2017]
```

```python
#confirm new subset
dfexamyear.head()
```

```python
#exclude other index numbers
dfexamyear1 =dfexamyear.iloc[:2]
dfexamyear1.drop(['Year'], axis=1)
```

```python
#confirm first two rows were seleted
dfexamyear1.head()
```

```python
dfexamyear1.shape
```

```python
#rename axis
dfexamyear1.set_index('School Name').plot.bar(rot=0, title='Core English Mean Score', figsize=(15,10), fontsize=12)
```

#Conclusion:
Two (2) Collegiate High Schools were chosen based upon type of school (Collegiate),a single Regents Exam type and year(Common Core English in 2017) with a similar number students taking the exam.  The graph respresents that both the Manhattan (62.8%) and Queens-based (68.5%) Collegiate schools tested with similar average mean test scores; though the Queens school had the higher mean average.
